```python
from collections import Counter

counts = Counter(ds["train"]["primary_subject"])

# View sorted by frequency
# for subject, count in counts.most_common():
#     print(f"{subject}: {count}")

# Or as a quick summary
print(f"\nTotal subjects: {len(counts)}")
print(f"Top 10: {counts.most_common(10)}")
print(f"Bottom 10: {counts.most_common()[-10:]}")
```

```python
from collections import Counter

all_subjects = []
for subjects_str in ds["train"]["subjects"]:
    all_subjects.extend(subjects_str.split("; "))

counts = Counter(all_subjects)
print(f"Total unique subjects: {len(counts)}")
print(counts.most_common(20))
```

```python
counts.most_common()[-10:]
```

```python
primary_subjects = set(ds["train"]["primary_subject"])
all_subjects = set(s for row in ds["train"]["subjects"] for s in row.split("; "))

never_primary = all_subjects - primary_subjects
print(f"Never a primary subject: {len(never_primary)}")
print(never_primary)
```
